<li> une fonction à 2 arguments (code postal, liste) capable de revoyer une liste des rues (une occurrence par rue) (++)    
<li> une fonction a 2 arguments (ville, liste) capable de renvoyer la distance maximale entre deux adresses de cette ville (++++)
<li> une fonction à 5 arguments (lat_n, lat_s, long_e,long_w,liste) capable de renvoyer toutes les addresses comprises dans ce cadre (+++)
<i> une fonction à 2 arguments (ville, liste) capable de renvoyer la densité urbaine de la ville (ratio: nb_adresse/cadran)

```python
import math 
def getPostal(code,liste):
    results=[]
    for a in liste:
        if a.split(',')[3]==code:
            results.append(a)
    return results

def getTown(ville,liste):
    results=[]
    for a in liste:
        if a.split(',')[4]==ville:
            results.append(a)
    return results

def getNorth(liste,nb_ville):
    results = liste[:nb_ville]
    for a in liste: 
        low_lat = min([l.split(',')[6] for l in results])
        lowest = [x for x in results if low_lat in x][0]
        
        if a.split(',')[6]> low_lat:
            results.append(a)
            results.remove(lowest)
    return results

def get_by_ZIP(code,liste,mode="ville"):
    results=[]
    for a in liste:
        if a.split(',')[3]==code:
            if mode == "ville":
                results.append(a.split(',')[4])
            elif mode == "rue":
                results.append(a.split(',')[2])
    return list(set(results))

def get_distance(p1,p2):
    ecart_long = abs(p1[0] - p2[0]) 
    ecart_lat = abs(p1[1] - p2[1])
    
    return ecart_long**2 + ecart_lat**2
    

def get_extremum(ville, liste):
    liste = [a for a in liste if a.split(',')[4]==ville]  # uniquement les adresses de la ville passée en argument
    
    distance_max = 0
    best_candidate = ()
    
    for a in liste:
        for b in liste:
            a_lat, a_long = float(a.split(",")[6]),float(a.split(",")[7])
            b_lat, b_long = float(b.split(",")[6]),float(b.split(",")[7])
            
            d = get_distance((a_lat, a_long),(b_lat, b_long))
            
            if d>distance_max:
                distance_max = d
                best_candidate = (a,b,d)
            
            #distances.append((a,b,d))
    return best_candidate


def cadran(min_lat, max_lat,min_long,max_long,liste):
    results = []
    
    for a in liste:
        #print(a)
        a_lat = float(a.split(",")[-2])
        a_long = float(a.split(",")[-1])
        if min_lat<a_lat<max_lat:
            if min_long<a_long<max_long:
                results.append(a)
    
    return results

def get_density(ville,liste):
    liste = [a for a in liste if a.split(',')[4]==ville]
    
    long = [float(a.split(',')[-1]) for a in liste]
    lat = [float(a.split(',')[-2]) for a in liste]
    
    long_min,long_max = min(long),max(long)
    lat_min,lat_max = min(lat),max(lat)
    
    a = (long_max-long_min)*(lat_max-lat_min)
    
    return len(liste)/a

def get_nb_town(liste): 
    return list(set([x.split(',')[4] for x in liste]))
```
